### Matplotlib example workflow

```python
x = [1,2,3,4]
y = [11, 22, 33, 44]
fig,ax = plt.subplots(figsize = (5,5))
ax.plot(x, y)
ax.set(title = 'Simple Plot', xlabel = 'x-axis', ylabel = 'y-axis')
```

### Make a plot from Dictionary

```python
nut_butter_prices = {'Almond butter':10 , 'Peanut butter': 8, 'cashew buter':12}
fig,ax = plt.subplots(figsize = (6,6))
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax.set(title = 'Parmar Butter Factory', xlabel = '-----Types of Butter-----', ylabel = '----- Price in $-----' );
```

```python
Goals_Scored = {'Barcelona': 80, 'Real Madrid': 75, 'Atletico Madrid':100 }
fig,ax = plt.subplots(figsize = (5,5))
ax.bar(Goals_Scored.keys(),Goals_Scored.values());
ax.set(title = 'LALIGA Goals Scored', xlabel = '******Name of club******', ylabel = '**No of goals scored in 20-21**');
```

### Histograms and Subplots

```python
marks_scored = {'Gurkirpal':100, 'Chandni': 33, 'Gurpreet':33}
fig,ax = plt.subplots(figsize = (5,5))
ax.barh(list(marks_scored.keys()), list(marks_scored.values()))
ax.set(title = 'Exam Result', xlabel='Candidate Marks', ylabel='Candidate Name' );
```

### Make some histogram

```python
x = np.random.randn(1000)
fig,ax = plt.subplots()
ax.hist(x);
```

```python
fig, ((ax1 , ax2 ), (ax3 , ax4)) = plt.subplots(nrows = 2, ncols =2, figsize = (10,5))
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

#### Another Method of making Subplots

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (10,5))
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```
