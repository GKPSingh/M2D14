### Matplotlib example workflow

```python
x = [1,2,3,4]
y = [11, 22, 33, 44]
fig,ax = plt.subplots(figsize = (5,5))
ax.plot(x, y)
ax.set(title = 'Simple Plot', xlabel = 'x-axis', ylabel = 'y-axis')
```

## Scatter Plot and Bar Plot

```python
x = np.linspace(0, 10, 100) 
fig,ax = plt.subplots()
y= x**2
ax.plot(x,y);
```

```python
x = np.linspace(0,10,100)
fig,ax =plt.subplots(figsize = (5,5))
y = x**2
ax.plot(x,y);
```

```python
x = np.linspace(0,10,100)
fig,ax = plt.subplots()
y = x**2
ax.scatter(x,y)
```

```python
x = np.linspace(0,10,100)
y = np.exp(x)
fig,ax = plt.subplots(figsize = (5,5))
ax.scatter(x,y);
```

```python
x = np.linspace(0,10,100)
fig,ax = plt.subplots(figsize = (6,6))
y = np.sin(x)
ax.scatter(x,y);
```

```python
x = np.linspace(0,10,100)
y = np.cos(x)
fig, ax = plt.subplots(figsize = (6,6))
ax.scatter(x,y)

```

```python
x = np.linspace(0,10,100)
y = np.tan(x)
fig, ax = plt.subplots(figsize = (6,6))
ax.scatter(x,y)
```

### Make a plot from Dictionary

```python
nut_butter_prices = {'Almond butter':10 , 'Peanut butter': 8, 'cashew buter':12}
fig,ax = plt.subplots(figsize = (5,5))
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax.set(title = 'Parmar Butter Factory', xlabel = 'Types of Butter', ylabel = 'Price in $' );
```
